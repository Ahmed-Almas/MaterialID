```python
import os
from ultralytics import YOLO
import cv2



# Update these paths with your uploaded files in Google Drive
VIDEOS_DIR = '/content/gdrive/My Drive/Prototypes/specs/videos'
video_path_out = '/content/gdrive/My Drive/Prototypes/specs/videos/r1_out.jpeg'

# Initialize the video capture object
cap = cv2.VideoCapture(VIDEOS_DIR)
ret, frame = cap.read()
H, W, _ = frame.shape
# Initialize the video writer
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'JPEG'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))

# Update the model path
model_path = '/content/gdrive/My Drive/Prototypes/specs/runs/detect/train/weights/best.pt'

# Load the YOLO model
model = YOLO(model_path)  # load a custom model

threshold = 0.5

while ret:
    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
